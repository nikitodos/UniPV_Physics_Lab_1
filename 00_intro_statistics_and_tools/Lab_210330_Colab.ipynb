### 1.3 Fitting

#### 1.3.1 - Definizione della funzione
E' la funzione che verrà usata per il fit. In questo caso particolare, la funzione è:
>$y = A + B\cdot t$

Si utilizza numpy perchè le variabili sono array di dati.


Modificare la cella sottostante per definire la funzione di fit per il nostro set di dati.

```python
def funzione(t,A,B): # Define fit function by its variables and parameters
  return A+B*t
```

#### 1.3.2 - Fit dei dati 
Si utilizza la funzione [curve_fit](https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.curve_fit.html) del modulo [scipy.optimize](https://docs.scipy.org/doc/scipy/reference/optimize.html).

A **curve_fit** vengono passati:
- la funzione da utilizzare per il fit definita in precedenza: **funzione()**
- gli array dei dati, cioè: t e d

E vengono ritornati:
- **popt**: l'array del parametri fittati
- **pcov**: la matrice di covarianza; i cui elementi della diagonale rappresentano la varianza dei rispettivi parametri. La radice quadrata dei quali da l'errore del parametro. 


```python
# Import curve_fit from cipy.optimize and fit using the previous function
from scipy.optimize import curve_fit
popt, pcov = curve_fit(funzione, Time, Distance)
```

```python
# Stampa dei parametri e degli errori
label = ["A", "B"]
for l, v, e in zip(label, popt, np.sqrt(np.diag(pcov))) :
  print("%10s = %9.6f +- %7.6f" % (l ,v, e))
```

###1.4 Plot della curva di fit


Disegnare la curva di fit, sovraimponendola ai dati.

```python
fig = plt.figure(figsize=(20,15))
ax = plt.axes()
plt.plot(Time, Distance, '.', label='Distance as a function of time')
plt.plot(Time, funzione(Time, popt[0], popt[1]), '-g', label='Fit function');
plt.title("Experimental values")
plt.xlabel("Time")
plt.ylabel("Distance")
plt.legend();
```

```python
import math
```

```python
# Define variables
data3 = np.genfromtxt('https://www2.pv.infn.it/~costanza/Potenza.txt', delimiter='\t')
lenght = data3[:,0]
period = data3[:,1]

# Define fit function by its variables and parameters
def pendulum(l,g):
  return 2*math.pi*np.sqrt(l/g)

# Fit using the previous function
popt3, pcov3 = curve_fit(pendulum, lenght, period)

# Define plot
plt.style.use('ggplot')
fig = plt.figure(figsize=(20,15))
ax = plt.axes()
plt.plot(lenght, period, '.', color='red', label='Period of a simple pendulum as a function of its length.')
plt.plot(lenght, pendulum(lenght, popt3[0]), '-g', label='Fit function');
plt.title("Fit function")
plt.xlabel("Pendulum lenght")
plt.ylabel("Period of oscillation")
plt.legend();
```

##Esercizio 4
1. Leggere il file di dati disponibile al seguente [link](https://www2.pv.infn.it/~costanza/Decadimento.txt). Il file contiene 2 colonne (tempo [giorni] e numero di nuclei $N(t)$), separate da una tab `\t`.
1. Fare il plot dei dati
1. Definire la funzione di fit\
$N(t)=N_0e^{-t/\tau}$
1. Fittare i dati
1. Disegnare la curva di fit, sovraimponendola al plot dei dati sperimentali.
1. Aggiungere titolo degli assi, titolo del grafico e legenda.


```python
# Define variables
data4 = np.genfromtxt('https://www2.pv.infn.it/~costanza/Decadimento.txt', delimiter='\t')
time4 = data4[:,0]
nuclei4 = data4[:,1]

# Define fit function by its variables and parameters
def decay(t, tao, N):
  return N*np.power(np.e, -t/tao)

# Fit using the previous function
popt4, pcov4 = curve_fit(decay, time4, nuclei4)

# Define plot
plt.style.use('ggplot')
fig = plt.figure(figsize=(20,15))
ax = plt.axes()
plt.plot(time4, nuclei4, '.', color='red', label='Number of radioactive nuclei as a function of time.')
plt.plot(time4, decay(time4, popt4[0], popt4[1]), '-g', label='Fit function');
plt.title("Fit function")
plt.xlabel("Time [d]")
plt.ylabel("Number of nuclei")
plt.legend();
```

##Esercizio 5
1. A partire dai dati sperimentali usati nell'esercizio 4, linearizzare la funzione esponenziale.
1. Disegnare il grafico delle variabili linearizzate
1. Fittare i dati.
1. Sovraimporre la curva di fit ai dati sperimentali.
1. Completare il grafico con titolo, titolo degli assi e legenda.

```python
# Define variables
nuclei5 = np.log(nuclei4)

# Define fit function by its variables and parameters
def linear_decay(t, C, D):
  return C+D*t

# Fit using the previous function
popt5, pcov5 = curve_fit(linear_decay, time4, nuclei5)

# Define plot
plt.style.use('ggplot')
fig = plt.figure(figsize=(20,15))
ax = plt.axes()
plt.plot(time4, nuclei5, '.', color='red', label='Linear decay of radioactive nuclei as a function of time.')
plt.plot(time4, linear_decay(time4, popt5[0], popt5[1]), '-g', label='Fit function');
plt.title("Fit function")
plt.xlabel("Time [d]")
plt.ylabel("Logarithmic number of nuclei")
plt.legend();
```
